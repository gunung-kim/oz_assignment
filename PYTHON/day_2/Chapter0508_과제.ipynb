## 모험가들의 레벨업 도전

### [배경 이야기]

당신은 마법사, 전사, 도적, 힐러 네 명의 모험가를 이끄는 리더입니다. 

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

### [동료 캐릭터와 능력치 정보]

masic : 힘(10), 민첩(20), 지능(80), 체력(50)

warrior : 힘(80), 민첩(30), 지능(20), 체력(100)

Assassin : 힘(40), 민첩(80), 지능(30), 체력(60)

healer : 힘(20), 민첩(40), 지능(70), 체력(70)

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다. 

마법을 사용하려면 지능이 50 이상이어야 합니다. 

마법사의 지능이 50 이상인지 확인하세요.

```python
#masic dict 생성
masic ={
    "힘" : 10,
    "민첩" : 20,
    "지능" : 80,
    "체력" : 50
}

#masic 지능 확인
answer=print(masic["지능"]>50)

#출력
print(answer)
```

<br>

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다. 

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
#warrior dict 생성
warrior = {
    "힘" : 80,
    "민첩" : 30,
    "지능" : 20,
    "체력" : 100
}

#warrior 힘, 체력 확인
answer_1=print(warrior["힘"]>=60)
answer_2=print(warrior["체력"]>=80)

#출력
print(answer_1,answer_2)
```

<br>

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다. 

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
# assassin dict 생성
assassin = {
    "힘" : 40,
    "민첩" : 80,
    "지능" : 30,
    "체력" : 60
}

#assassin 민첩, 힘 확인
answer_3=print(assassin["민첩"]>=70)
answer_4=print(assassin["힘"]>=50)

#출력
print(answer_3,answer_4)
```

<br>

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다. 

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
#healer dict 생성
healer = {
    "힘" : 20,
    "민첩" : 40,
    "지능" : 70,
    "체력" : 70
}

#healer 지능,체력 확인
answer_5=print(healer["지능"]>=60)
answer_6=print(healer["체력"]>=50)

#출력
print(answer_5,answer_6)
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다. 


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
#각 dict에 경험치(key)와 경험치양(values) 추가
masic["경험치"] = 120
warrior["경험치"] = 150
assassin["경험치"] = 110
healer["경험치"] = 90

#경험치 100이상인지 확인 및 출력 
print(masic["경험치"]>=100)
print(warrior["경험치"]>=100)
print(assassin["경험치"]>=100)
print(healer["경험치"]>=100)
```
