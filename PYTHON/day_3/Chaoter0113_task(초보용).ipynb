__[문제 2: 입금]__

가지고 있는 돈을 입금합니다.

입금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 입금 거래를 영수증 리스트에 기록하세요.

- 입금 금액은 input()을 이용해 받습니다.
- 영수증은 list()로 선언합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.

(변수명 예시: 영수증 리스트 - receipts, 입금 금액 - deposit_amount)


```python

#입금할 금액/출금할 금액 입력받기
plus_money = int(input("압금하실 금액을 입력해주세요"))
#계좌 잔액이랑 더한 후 변수 다시 할당
my_account += plus_money
#현재 잔액 표기
print(my_account)
#영수증
recepits = {}

#영수증에 기록
recepits["plus"] = [plus_money]
print(recepits)
```

__[문제 3: 출금]__

사용자가 돈을 출금합니다. 출금 금액을 입력받고, 잔액을 갱신한 후 새로운 잔액을 출력하세요. 출금 거래를 영수증 리스트에 기록하세요. 

- 출금 금액은 input()을 이용해 받습니다.
- 현재 잔액보다 출금 요청 금액이 많을 경우 현재 잔액 만큼만 출금되도록 합니다.
- 영수증에 들어가는 정보는 튜플 형태입니다.
- 튜플에 담기는 정보는 (입금 or 출금, 입금 or 출금 금액, 현재 잔액)순 입니다.

(변수명 예시: 출금금액 - withdraw_amount)

```python
#계좌생성 - 초기 금액은 1,000원
my_account = 1000
print(my_account)


#입금할 금액/출금할 금액 입력받기
plus_money = int(input("압금하실 금액을 입력해주세요"))
minus_money = int(input("인출하실 금액을 입력해주세요"))

#계좌 잔액이랑 더한 후 변수 다시 할당
my_account += plus_money

#현재 잔액 표기
print(my_account)
#영수증
recepits = {}

#영수증에 기록
recepits["plus"] = [plus_money]
print(recepits)

#출금하기
if my_account >= minus_money :
    pass
elif my_account < minus_money :
    minus_money = my_account
    print(f"현재 잔액은 {my_account}입니다. 요청주신 금액보다 {minus_money - my_account} 부족합니다.")
    ask = input("계좌잔액 전체를 출금할까요? y/n")
    if ask =='y':
        minus_money = my_account
    elif ask == 'n':
        minus_money = 0
        print('출금을 취소합니다.') 

my_account -= minus_money
recepits["minus"] = [minus_money]

print(recepits)
```
