[문제 2] 여러 숫자를 입력받아 그 합계를 반환하는 함수 sum_numbers를 작성하세요. 이 때, 가변 매개변수를 사용하세요.

```python
#여러숫자를 입력 받아 -> input과 가변 매개변수(*) 사용
#입력된 숫자들의 합 -> sum 사용

def sum_numbers(*args:int)->int:
    """여러숫자를 입력받아 입력된 숫자들의 합을 구하는 함수"""
    a = args    #튜블 언패킹
    b=[]
    for i in a:
        c=int(i)
        b.append(c)
    return sum(b)

n=input("숫자입력").split(",")
sum_numbers(*n) 
```

[문제 8] 사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
#사용자로부터 3개의 숫자 입력받고
#튜플로 변환하여 최솟값과 최대값을 구한다
def min_max(*args):
    a = args
    args_int_list =[int(b) for b in a] #str로 입력된 값들을 int로 형변환 후 리스트로 감싸기
    args_int_tuple = (args_int_list) #다시 튜플로
    최소 = min(args_int_tuple)
    최대 = max(args_int_tuple)
    return 최소,최대
n =input("숫자입력").split() 
min_max(*n)    
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string (s1:str,s2:str)->str:
    """두개의 문자열을 입력받아 두 문자열의 길이를 비교하고 길이가 긴 문자열을 반환하는 함수 
    문자열이 같을 경우 첫번째로 입력했던 문자열이 반환"""
    if len(s1) < len(s2):
        return s2
    else :
        return s1
    
s1,s2 = input("두개의 문자열을 '/'기준으로 입력하세요").split('/')
longer_string(s1,s2)
```
