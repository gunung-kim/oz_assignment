1478 → "one4seveneight"  
234567 → "23four5six7"

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나,

s가 의미하는 원래 숫자를 출력하도록 코드를 작성해주세요

s = "77three4one

참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

숫자	영단어

6	six  
7	seven  
8	eight  
9	nine

__[문제4]__

행렬의 덧셈은 행과 열의 크기가 같은 두 행렬의 같은 행, 같은 열의 값을 서로 더한 결과가 됩니다. 

 행렬 arr1과 arr2가 주어졌을 때, 행렬 덧셈의 결과를 출력하는 코드를 작성해주세요(result와 같은 결과가 출력되도록 작성)

arr1 = [[1,2],[2,3]]

arr2 = [[3,4],[5,6]]

result = [[4,6],[7,9]]

```python
arr1 = [[1,2],[2,3]]
arr2 = [[3,4],[5,6]]
#각각 리스트의 

result = []#답 넣을 리스트 만들기
for a,b in zip(arr1,arr2):#각리스트의 각 행을 a,b에 대임
    v=[]#각 a,b에서의 각 열의 합을 담기위한 리스트
    for c,d in zip(a,b):#각 a,b의 각 열을 c,d에 대입하고 더하기
        v.append(c+d)#리스트에 담기
    result.append(v)#답 리스트에 담기
print(result)

'''
a = arr1[0][0]+arr2[0][0]
b = arr1[0][1]+arr2[0][1]
c = arr1[1][0]+arr2[1][0]
d = arr1[1][1]+arr2[1][1]
resuit = [[a,b],[c,d]]
print(resuit)
'''
#조교님 정답
result = []#결과값을 받을 리스트
for a,b in zip(arr1,arr2):#arr1과 arr2의 행1,행2를 각각 a,b 대입 
    re = [c+d for c,d in zip(a,b)]#a,b에 대입된 각 행의 열1열2를 c,d에 대입하고 더한후 re에 추가
    result.append(re)#re를 result에 추가
print(result)
```

String형 배열 seoul의 element중 "Kim"의 인덱스 x를 찾아, "김서방은 x에 있다"는 String을 출력하는 코드를 작성해주세요.   

seoul에 "Kim"은 오직 한 번만 나타나며 잘못된 값이 입력되는 경우는 없습니다.

seoul = ["Jane", "Kim"]

출력 예) "김서방은 1에 있다"

```python
seoul = ["jane","park","kim"]

for i in seoul :#i에 seoul값들 대입
    if i == "kim" :#kim 찾기
        a = seoul.index(i)#kim 인덱스 찾기
        print(f"김서방은 {a}에 있다")#kim 인덱스 값 넣기
```
