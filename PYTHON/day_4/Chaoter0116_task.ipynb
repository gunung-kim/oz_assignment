- 현재 붕어빵의 재고를 딕셔너리로 관리합니다.
- 딕셔너리에는 붕어빵의 종류와 각각의 재고 개수가 저장됩니다.

```python
'''주문을 받고 주문이 들아오면 계산 재고관리를 바로해야함
관리자 모드를 따로만들어서 관리 할수 있야한다'''
'''-----------------------구상도----------------------------
    1. 관리자    -   메뉴 추가하기   - 메뉴 수정하기
                 -  재고관리하기 
                 -  아무나 못 들어가게 비밀번호 걸어두기
    2. 소비자     - 주문하기
                 - 메뉴 선택 후 가격 및 갯수 확인
                 - 결제
-------------------------------------------------------------'''
#전체 dict 만들고 안에 서브 dict 넣기
#전체 dict의 key값은 "팥붕","슈붕","초붕" 각 key의 vlaue는 서브dict 이름
#서브 dict의 key값은 "가격","재고","판매횟수","주문수"
#서브 dict 생성
key2 = ["팥붕","슈붕","초붕"]
value = ["가격","재고","판매횟수","주문수"]
value2 = [0,0,0,0]

팥붕 = {a:b for a,b in zip(value,value2)}
슈붕 = {a:b for a,b in zip(value,value2)}
초붕 = {a:b for a,b in zip(value,value2)}

붕어빵 = {
    "팥붕" : 팥붕,
    "슈붕" : 슈붕,
    "초붕" : 초붕
}
#재고 및 가격 추가
붕어빵["팥붕"]["가격"] = 1000
붕어빵["팥붕"]["재고"] = 10
붕어빵["슈붕"]["가격"] = 1500
붕어빵["슈붕"]["재고"] = 8
붕어빵["초붕"]["가격"] = 2000
붕어빵["초붕"]["재고"] = 5

print(붕어빵)
print(팥붕)
print(붕어빵["팥붕"]["재고"])
print(붕어빵["슈붕"]["재고"])
print(붕어빵["초붕"]["재고"])
```

__[문제 3: 붕어빵 주문 처리]__

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
key2 = ["팥붕","슈붕","초붕"]
value = ["가격","재고","판매횟수"]
value2 = [0,0,0]

팥붕 = {a:b for a,b in zip(value,value2)}
슈붕 = {a:b for a,b in zip(value,value2)}
초붕 = {a:b for a,b in zip(value,value2)}

붕어빵 = {
    "팥붕" : 팥붕,
    "슈붕" : 슈붕,
    "초붕" : 초붕
}
#재고 및 가격 추가
붕어빵["팥붕"]["가격"] = 1000
붕어빵["팥붕"]["재고"] = 10
붕어빵["슈붕"]["가격"] = 1500
붕어빵["슈붕"]["재고"] = 8
붕어빵["초붕"]["가격"] = 2000
붕어빵["초붕"]["재고"] = 5

loop_1 = True
while loop_1:
    주문1,주문2 = input("원하시는 붕어빵과 수를 입력해주세요 ex)팥붕,3").split(',')
    재고=붕어빵[주문1]["재고"]
    if int(주문2) > 붕어빵[주문1]["재고"]:
        print(f"재고가 부족합니다 {재고}개 이하로 주문해주세요")
    elif int(주문2) < 붕어빵[주문1]["재고"]:
        결제금액 = 붕어빵[주문1]["가격"]*int(주문2)
        print(f"{주문1} {주문2}개 주문이 완료되었습니다")
        print(f"결제금액은 {결제금액}입니다")
        붕어빵[주문1]["재고"] -= int(주문2)
        붕어빵[주문1]["판매횟수"] += 1
        loop_1 = False
print(붕어빵[주문1]["재고"])
print(붕어빵[주문1]["판매횟수"])    
 
```
