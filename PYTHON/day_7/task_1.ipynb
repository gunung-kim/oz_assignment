- 입력 중 숫자가 아닌 값이 입력되면 ValueError를 처리
- 숫자 외 값이 입력된 경우 어떤 값이 잘못 입력되었는지 사용자에게 알려주는 문구 출력<br>
  (예: f"잘못된 값: {변수}. 숫자를 입력해주세요.")



````python
save = []
while 1:
    try:
        user_input = input("숫자를 입력하세요").split(" ")
        if "done" in user_input :
            break
        else :
            save_1 = list(map(int,user_input))
            save.append(save_1)
    except ValueError :
        print(f"잘못된 값 : {user_input}")        


````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요. 

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time
def timer_decorator(function):
    def warpper(*args,**kwargs):
        start_time = time.time()
        result = function(*args,**kwargs)
        end_time = time.time()
        a = end_time - start_time
        print(f"실행시간은 {a}초 입니다.")
        return result
    return warpper

@timer_decorator
def func(a,b):
    return a+b

func(10348170,4301740)
````

```
animals/(폴더)
```

````python
f=open('animaa')
class animals:
    def __init__(self,name,food,sound):
        self.name = name
        self.food = food
        self.sound = sound



````

````python
from animals import animals

class mammals(animals):
    def info(self):
        print(f"{self.name}은 포유류이고 주식은 {self.food}이며 {self.sound} 소리를 냅니다")
dog = mammals("깜식이","사료","멍멍")
````
